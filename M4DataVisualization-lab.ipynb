```python
res = cur.execute("SELECT name FROM sqlite_schema WHERE type = 'table'")
res.fetchall()
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT ConvertedComp FROM master
"""

df = pd.read_sql_query(QUERY, conn)
plt.hist(df["ConvertedComp"], bins = 15)
plt.xlabel("Salary Range")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age FROM master
"""

df = pd.read_sql_query(QUERY, conn)
cleaned = df["Age"].dropna()
plt.boxplot(cleaned)
plt.ylabel("Age")
plt.title("Age Distribution")
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""

df = pd.read_sql_query(QUERY, conn)
plt.scatter(x = df["WorkWeekHrs"], y = df["Age"])
plt.xlabel("Work Week Hours")
plt.ylabel("Age")
plt.title("Age vs, Work Week Hours")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import seaborn as sns
```

```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""

df = pd.read_sql_query(QUERY, conn)
sns.scatterplot(data = df, x = "WorkWeekHrs", y = "CodeRevHrs", size = "Age")
plt.show()

```

Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.
